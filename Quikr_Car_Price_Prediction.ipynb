## Data Cleaning

```python
# Problems noticed :
# year has so many non year values
# some price has non-numeric values(Ask For Price)
# covert price and year to int by removing comma (object to int)
# kms_driven has object values with kms at last.
# It has nan values and some non km values
# fuel_type has nan values
# Keep first three words of the name
```

```python
# year has many non-year values
df = df[df['year'].str.isnumeric()]
```

```python
# year is in object --> Change to integer
df['year'] = df['year'].astype(int)
```

```python
# Price has Ask for Price
df = df[df['Price']!='Ask For Price']
```

```python
# Price has commas in its prices and is in object
df['Price'] = df['Price'].str.replace(',','').astype(int)
```

```python
# kms_driven has object values with kms at last.
df['kms_driven'] = df['kms_driven'].str.split().str.get(0).str.replace(',','')
```

```python
# It has nan values and some non km values
df = df[df['kms_driven'].str.isnumeric()]
```

```python
# Change to integer
df['kms_driven'] = df['kms_driven'].astype(int)
```

```python
df.isnull().sum()
```

```python
#fuel_type has nan values
df=df[~df['fuel_type'].isna()]
```

```python
df.info()
```

```python
# Keep first three words of the name
df['name'] = df['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

```python
# Resetting the index of the final cleaned data
df = df.reset_index(drop=True)
```

## Data Preprocessing

```python
x = df.drop(['Price'], axis=1)
y = df['Price']
```

```python
x.head()
```

```python
y.head()
```

```python
#One-hot-encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])

ct = make_column_transformer((OneHotEncoder(categories = ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x ,y, test_size= 0.2, random_state = 0)

print("Training set size :", len(x_train))
print("Test set size :", len(x_test))
```

## Model Training and Evaluation

```python
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
# Random Forest Regression Model
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators= 100, criterion='squared_error',max_features = 'log2', min_samples_split= 4,n_jobs =-1, random_state = 0)

pipe = make_pipeline(ct,rf)

pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2 = r2_score(y_test,y_pred)
r2 = round(r2, 2)
print("R^2 score :", r2)
```

```python
# XGBoost Regression Model
import xgboost as xgb
xgbr = xgb.XGBRegressor(learning_rate = 0.1, n_estimators = 550, objective = "reg:squarederror",reg_lambda = 1.5, max_depth = 2,n_jobs =-1, random_state = 0)
pipe = make_pipeline(ct,xgbr)
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2 = r2_score(y_test,y_pred)
r2 = round(r2, 2)
print("R^2 score :", r2)
```
